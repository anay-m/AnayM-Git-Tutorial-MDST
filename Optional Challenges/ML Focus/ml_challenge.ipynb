```python
import matplotlib.pyplot as plt
import numpy as np
```

## Dataset

```python
# DO NOT MODIFY
from sklearn.datasets import make_moons

X, Y = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)
```

## Training

```python
f = plt.figure(figsize=(12,6))
plt.scatter(X[:,0], Y, color = "blue")
plt.scatter(X[:,1], Y, color = "black")
#Plots a scatter plat depicting the relationship between each feature of X and Y.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.3, random_state=0)
#Splits data into testing and training data by 70/30 percent split

from sklearn import linear_model
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
#Uses logistic regression to address binary classification problem
```

## Testing / Performance Analysis

```python
model_weights = model.coef_
model_bias = model.intercept_
print("Model Weight: " + str(model_weights))
print("Model Bias Term: " + str(model_bias))
#Printing coefficients and bias term of regression model

y_train_predictions = model.predict(X_train)
y_test_predictions = model.predict(X_test)

mean_abs_err_train = np.mean(np.abs(y_train_predictions-y_train))
mean_abs_err_test = np.mean(np.abs(y_test_predictions-y_test))
root_mean_sqr_err_train = np.mean((y_train_predictions-y_train)**2)**0.5
root_mean_sqr_err_test = np.mean((y_test_predictions-y_test)**2)**0.5
print("Mean absolute error on training set: " + str(mean_abs_err_train))
print("Root mean square error on training set: " + str(root_mean_sqr_err_train))
print("Root mean square error on test set: " + str(root_mean_sqr_err_test))
print("Mean absolute error on test set: " + str(mean_abs_err_test))
#Printing error calculations to analyze model performance
```

```python
f = plt.figure(figsize=(12,6))
plt.scatter(X_train[:,1], y_train, color = "red")
plt.scatter(X_train[:,1], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.title("Model Behavior On Training Set")
#Plotting the model predictions and actual output values of training set
```

```python
f = plt.figure(figsize=(12,6))
plt.scatter(X_test[:,1], y_test, color = "blue")
plt.scatter(X_test[:,1], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.title("Model Behavior on Testing Set")
#Plotting the model predictions and actual output values of test set
```
